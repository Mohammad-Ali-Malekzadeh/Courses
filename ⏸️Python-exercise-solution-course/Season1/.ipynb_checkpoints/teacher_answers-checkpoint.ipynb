1. عددی را از کاربر دریافت کنید و در خروجی زوج بودن یا نبودن آنرا نمایش دهید.

```python
num = int(input("Ye Adad Vared Konid: "))
if num % 2 == 0:
    print(num, 'is even')
else:
    print(num, 'is odd')
```

2. عددی بزرگتر از صفر را از کاربر دریافت کنید و با استفاده از  while، از آن عدد تا صد را نمایش دهید. 

```python
num = int(input("Ye Adad Vared Konid: "))
while num <= 100:
    print(num)
    num += 1
```

3. سه عدد را از کاربر  گرفته و ماکزیمم آنها را برگردانید.

```python
n1 = 11
n2 = 10
n3 = 3
if n1 > n2 and n1 > n3:
    print(n1)
elif n2 > n1 and n2 > n3:
    print(n2)
else:
    print(n3)
```

4. یک عدد مثبت را از کاربر گرفته و مجمو ع تمام اعداد بین 1  و آن عدد را محاسبه کنید و نمایش دهید.

```python
a = int(input("Enter an integer: "))
s = 0
for i in range(1, a+1):
    s += i
print(s)
```
